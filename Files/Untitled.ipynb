```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
K = 50  # Example strike price
S_T = np.linspace(0, 100, 500)  # Stock prices from 0 to 100
V_T = np.maximum(S_T - K, 0)  # Call option value at expiration

# Creating the plot
plt.figure(figsize=(8, 5))
plt.plot(S_T, V_T, color='red', label='$c_T = \max(S_T - K, 0)$', zorder=3)
plt.axhline(0, color='black', linewidth=1, zorder=2)  # Zero value line without a legend
plt.title('Value at Expiration', fontsize=16)
plt.xlabel('$S_T$', fontsize=14)
plt.ylabel('$V_T$', fontsize=14)
plt.axvline(x=K, color='grey', linestyle='--', zorder=1)  # Marking the strike price K
plt.text(K, -3, '$K$', fontsize=14, color='grey', verticalalignment='top', horizontalalignment='center')  # Lowering K below the horizontal line at zero

plt.axis('square')  # Makes the plot square
plt.ylim(None, 60)

plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define strike prices for the iron condor
K1, K2, K3, K4 = 30, 40, 60, 70

# Define a range for the stock price at expiration
S_T = np.linspace(0, 100, 500)

# Calculate payoffs for the put options
put_long = np.maximum(K1 - S_T, 0)
put_short = -np.maximum(K2 - S_T, 0)

# Calculate payoffs for the call options
call_long = np.maximum(S_T - K4, 0)
call_short = -np.maximum(S_T - K3, 0)

# Total profit calculation from the iron condor
V_Total = put_long + put_short + call_long + call_short

# Set up the plot
plt.figure(figsize=(10, 6))
plt.plot(S_T, V_Total + 6, label='Iron Condor Payoff', color='black')

# Add lines and labels to indicate strike prices
for K, label in zip([K1, K2, K3, K4], ['$K_1$', '$K_2$', '$K_3$', '$K_4$']):
    plt.axvline(x=K, color='grey', linestyle='--', zorder=1)
    plt.text(K, -21.5, label, fontsize=18, color='grey', verticalalignment='top', horizontalalignment='center')

# Decorate the plot
plt.title('Iron Condor Profit at Expiration', fontsize=16)
plt.xlabel('$S_T$', fontsize=16)
plt.ylabel('$\Pi_T$', fontsize=16)
plt.axhline(0, color='black', linewidth=1, zorder=2)  # Zero profit line
plt.grid(True)


# Adding text annotations directly in the plot
plt.text(50, 8, '$\Pi_T$', fontsize=14, color='black')

plt.axis('square')  # Makes the plot square

plt.ylim(-20, 20)
#plt.legend()
# Suppress x-axis numbers

plt.gca().set_xticklabels([])

plt.show()
```
